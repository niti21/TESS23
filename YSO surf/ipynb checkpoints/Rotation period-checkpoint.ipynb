```python
import lightkurve as lk
import matplotlib.pyplot as plt

# List of TIC IDs
tic_ids = [ 64837857, 249784843, 388696341, 428274538, 361944360, 66539637, 64077901]

# Sector number
sector_number = 17

# Function to find the rotation period
def find_rotation_period(tic_id, sector):
    # Search and download TESS light curve for the given TIC ID and sector
    search_result = lk.search_lightcurve(f'TIC {tic_id}', mission='TESS', sector=sector)
    lc = search_result.download().PDCSAP_FLUX

    # Plot the light curve
    lc.plot()
    plt.title(f'Light Curve - TIC {tic_id} (Sector {sector})')
    plt.xlabel('Time')
    plt.ylabel('Flux')
    plt.show()

    # Estimate the rotation period using the Lomb-Scargle periodogram
    periodogram = lc.to_periodogram()
    rotation_period = periodogram.period_at_max_power

    return rotation_period

# Loop through the TIC IDs and find the rotation period for each
for tic_id in tic_ids:
    rotation_period = find_rotation_period(tic_id, sector_number)
    print(f'TIC {tic_id}: Rotation Period = {rotation_period:.2f} days')
```
